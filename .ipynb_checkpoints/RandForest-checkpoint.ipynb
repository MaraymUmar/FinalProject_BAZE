```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("dataSets.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df1=df.drop(columns=["Food_Items"])
df1.head()
```

```python
series = df['Calories']
df_single_column = series.to_frame()
```

```python
#normalizing the calories column
scaler = MinMaxScaler()
scaler.fit(df1[['Calories']])
df1['Calories'] = scaler.transform(df1[['Calories']])
df1
```

```python
##Encoding
df.Glycemic_Index.unique()
```

```python
le = LabelEncoder()
df1["Glycemic_Index"]=le.fit_transform(df["Glycemic_Index"])
df1.head()
```

```python
df1.Sodium_Content.unique()
```

```python
le = LabelEncoder()
df1["Sodium_Content"]=le.fit_transform(df1["Sodium_Content"])
df1.head()
```

```python
df1.Veg_Nonveg.unique()
```

```python
le = LabelEncoder()
df1["Veg_Nonveg"]=le.fit_transform(df1["Veg_Nonveg"])
df1.head()
```

```python
y= df1["Glycemic_Index"]
y.head()
```

```python
X=df1.drop(columns=["Contains_Nuts", "Contains_Glutten","Contains_Diary", "Veg_Nonveg", "Protein", "Iron", "Calcium", "Sodium", "Potassium", "Fibre", "Magnesium", "Zinc", "Carbohydrates"])
X.head()
```

```python
##Splitiing of the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
print("X_train", X_train.shape)
print("X_test", X_test.shape)
```

```python
##Random Forest Classification on training set
from sklearn.ensemble import RandomForestClassifier
cclf = RandomForestClassifier(n_estimators = 5, criterion = "gini")
cclf.fit(X_train, y_train)
```

```python
y_pred = cclf.predict(X_test)
y_pred
```

```python
##Confucion Matrix of the Model
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
